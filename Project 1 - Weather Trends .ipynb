```python
#Explore global_temp
global_temp.plot.line(x = 'year', y ='avg_temp', title = 'Average Temperature Globally', xlabel = 'Recording period (25 Year increments)', ylabel = 'Temperature (C)');
```

```python
# Combine the data together to get a better analysis of the temp data

from matplotlib import pyplot as plt
plt.figure(figsize=(20,10))
plt.plot( 'year', 'avg_temp', data=seoul_temp, color='red')
plt.plot( 'year', 'avg_temp', data=global_temp, color='blue')
plt.xlabel('Recording Period in Years (50 Years Increments)', fontsize = 25)
plt.ylabel('Temperature (C)', fontsize = 25)
plt.legend(['Seoul', 'Global'])
plt.title('Moving Average: Seoul Temperature VS Global Temperature', fontsize = 25)
plt.show()
```

```python
# Import Libraries 
import pandas as pd
import seaborn as sns
```

```python
# Import Data
seoul_temp = pd.read_csv('seoul data.csv')
```

```python
global_temp = pd.read_csv('global.csv')
```

```python

seoul_temp.count()
```

```python
#Explore seoul_temp data
seoul_temp.plot.line(x = 'year', y ='avg_temp', color = 'red')
```

```python
#Explore global_temp
global_temp.plot.line(x = 'year', y ='avg_temp')
```

```python
# Combine the data together to get a better analysis of the temp data

from matplotlib import pyplot as plt
plt.figure(figsize=(20,10))
plt.plot( 'year', 'avg_temp', data=seoul_temp, color='red')
plt.plot( 'year', 'avg_temp', data=global_temp, color='blue')
plt.xlabel('Years (50 Years period)', fontsize = 25)
plt.ylabel('Average Temperature', fontsize = 25)
plt.legend(['Seoul', 'Global'])
plt.show()
```
